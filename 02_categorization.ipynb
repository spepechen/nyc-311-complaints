#K-means clustering for complaints 

```python
vectorizer = CountVectorizer()
data = vectorizer.fit_transform(complaints_list).toarray()
```

```python
number_of_clusters = 20
kmeans = KMeans(n_clusters=number_of_clusters)
kmeans.fit(data)
```

```python
def clusters_to_csv(labels, types):
    '''
    Helper function to turn scikit-learn clusters into abbreviated CSVs (Chase's)
    '''
    for k in set(labels):
        class_members = [index[0] for index in np.argwhere(labels == k)]
        for index in class_members:
            print '%s,%s' % (int(k), types[index])
```

```python
clusters_to_csv(kmeans.labels_, complaints_list)
```

###Getting one-sentence-summary of each cluster

```python
def clusters_to_list_of_dict(labels, types):
    '''
    Helper function to turn abbreviated CSVs into list of dictionaries
    '''
    clusters_list=[]
    for k in set(labels):
        class_members = [index[0] for index in np.argwhere(labels == k)]
        for index in class_members:
            #s =(int(k), types[index])  
            temp = {}
            temp['cluster_no']= int(k)
            temp['content']= types[index]
            clusters_list.append(temp)
    return clusters_list
```

```python
from collections import Counter

def clusters_summary(labels, types ,num):
    
    '''
    Helper function to turn list of dictionaries into a summary of the cluster
    OUPUT: 
    index of the cluster 
    size of the cluster 
    top 3 word frequency of the cluster
    '''
    clusters = clusters_to_list_of_dict(labels, types)
    cluster1_list = [c['content'] for c in clusters if c['cluster_no'] == num]
    
    cluster_index = num
            
    size = len(cluster1_list)
                
    cluster_str = " ".join(cluster1_list)
    labels = Counter(cluster_str.split()).most_common()[:5]
                
    return cluster_index, size, labels
```

```python
clusters = clusters_to_list_of_dict(kmeans.labels_, complaints_list)
clusters 
```

```python
def classifier(x): 
    # CLASS 1: UTILITIES
    utilities_list =['Drinking Water',
                     'ELECTRIC',
                     'General Construction/Plumbing',
                     'HEAT/HOT WATER',
                     'HEATING',
                     'PAINT - PLASTER',
                     'PAINT/PLASTER',
                     'PLUMBING',
                     'Sewer',
                     'WATER LEAK',
                     'Water System']
    # CLASS 2: TRAFFIC
    traffic_wordlist =['Blocked Driveway',
                        'Broken Muni Meter',
                        'Illegal Parking',
                        'Street Condition',
                        'Traffic Signal Condition']
    # CLASS 3: NOSIE
    noise_wordlist =['Noise',
                     'Noise - Commercial',
                     'Noise - Helicopter',
                     'Noise - House of Worship',
                     'Noise - Park',
                     'Noise - Street/Sidewalk',
                     'Noise - Vehicle',
                     'Noise Survey']
    # CLASS 4: TRAFFIC
    sanitary_wordlist =['Dirty Conditions',
                         'Dirty Conditions',
                         'Overflowing Recycling Baskets',
                         'Rodent',
                         'Sanitation Condition',
                         'UNSANITARY CONDITION',
                         'Unlicensed Dog',
                         'Unsanitary Animal Facility',
                         'Unsanitary Animal Pvt Property',
                         'Unsanitary Pigeon Condition',
                         'Urinating in Public']

    if x in utilities_list:
        return "utilities_problem"
    elif x in traffic_wordlist:
        return "traffic_problem"
    elif x in noise_wordlist:
        return "noise_problem"
    elif x in sanitary_wordlist:
        return "sanitation_problem"
    else: 
        return "others"
```

```python
df['Complaint Classification'] = df['Complaint Type'].apply(lambda x : classifier(x))
```
